<h2>Section 5: Data Science

Question:
Do players who play more minutes per game (MP) tend to score more points per game (PTS)?

Why this question:
In Section 4, I noticed that some players with higher playing time seemed to have higher scoring performance. I wanted to test whether this pattern holds across the dataset using a predictive model and quantify the strength of the relationship.

How I answered it using data science:
To answer this question, I used a linear regression model to predict players' points per game (PTS) based on their minutes per game (MP). Linear regression is a machine learning technique that helps model the relationship between independent and dependent variables. After training the model, I predicted the PTS values and calculated the RMSE (root mean squared error) to evaluate model performance.

Results:

The correlation between MP and PTS was 0.426, indicating a moderate positive relationship.
The linear regression model produced an RMSE of 2.53, which suggests the model was moderately accurate in predicting scoring performance based on playing time.
This implies that playing time contributes to scoring potential, but other factors (e.g., shooting efficiency, player role) are also influential.

```python
import pandas as pd
df = pd.read_csv("nba_top_50_players_2025.csv")
correlation = df['MP'].corr(df['PTS'])
print("Correlation between MP and PTS:", correlation)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
ind = ["MP"]
dep = ["PTS"]
model2= model.fit(df[ind], df[dep])
model.predict(df[ind])
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_true = df[dep]
y_pred = model2.predict(df[ind])

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
rmse
```

<h2> Section 6: A Different Data Visualization

Explanation:
In this section, I created a box plot to show how scoring performance (PTS) varies across different player positions (e.g., PG, SG, SF, PF, C). This visualization differs from the scatter plot used in Section 3 and is effective for comparing distributions.

From the box plot, we can observe:

Guards (PG, SG) tend to have more variability in scoring.

Forwards and Centers (SF, PF, C) generally score within a narrower range.

Some outliers indicate exceptionally high scorers in each group.（e.x. Alexander and Jokic has Significantly stronger scoring ability than other players in the same position)

This shows how player role (position) impacts their scoring output, possibly due to differences in team roles and usage rates.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot of PTS grouped by position
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Pos', y='PTS')
plt.title('Distribution of Points per Game by Player Position')
plt.xlabel('Player Position')
plt.ylabel('Points per Game (PTS)')
plt.show()
```
